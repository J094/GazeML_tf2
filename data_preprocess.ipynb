```python
import tensorflow as tf
```

```python
batch_size = 32
eyes = []
label_ldmks = []
label_radius = []
for i in range(batch_size):
    entry = next(u.entry_generator())
    while(entry is None):
        entry = next(u.entry_generator())
    eyes.append(entry['eye'])
    label_ldmks.append(entry['landmarks'])
    label_radius.append(entry['radius'])
eyes = tf.constant(eyes)
label_ldmks = tf.constant(label_ldmks)
label_radius = tf.constant(label_radius)
```
